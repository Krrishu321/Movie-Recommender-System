# **MISSING VALUE IMPUTATION**

```python
#we are checking the high percentage of missing value in a dataset
round(100*(data.isnull().sum()/len(data.index)),2)
```

we are see "gross" and "budget" columns have large number of NaN values, drop all the rows with NaNs at this columns using the

## **ISNAN FUNCTION OF NUMPY ALONG WITH A NEGATION(~)**

```python
data = data[~np.isnan(data['gross'])]
data = data[~np.isnan(data['budget'])]
```

```python
data.isnull().sum()
```

```python
#the rows for which the sum of Null is less than 2 are retained
data = data[data.isnull().sum(axis=1) <=2]
data.isnull().sum()
```

# WE ARE IMPUTTING THE MISSING VALUES

```python
#USING MEAN FOR NUMERICAL COLUMNS

data['num_critic_for_reviews'].fillna(data['num_critic_for_reviews'].mean(), inplace = True)
data['duration'].fillna(data['duration'].mean(), inplace = True)


#USING MODE FOR CATEGORICAL COLUMNS

data['language'].fillna(data['language'].mode()[0], inplace = True)


#AS WE KNOW THAT WE CAN NOT USE STATISTICAL VALUES FOR IMPUTING THE MISSING VALUES OF ACTOR NAMES, SO WE WILL REPLACE THE
## ACTOR NAMES WITH "UNKNOWN ACTOR"

data['actor_2_name'].fillna('Unknown Actor', inplace = True)
data['actor_3_name'].fillna('Unknown Actor', inplace = True)
data['actor_1_name'].fillna('Unknown Actor', inplace = True)


#WE are checking all the missing values imputed
#lets check the no of total missing values in dataset


data.isnull().sum().sum()


```

# ***FEATURE ENGINEERING***

```python
#LETS CONVERTS THE GROSS AND THE BUDGET FROM $ TO MILLION $ TO MAKE OUR ANALYSIS EASIER

data['gross'] = data['gross']/1000000
data['budget'] = data['budget']/1000000

```

```python
# LETS CREATE THE PROFIT COLUMN USING THE GROSS AND BUDGET COLUMNS

data['Profit'] = data['gross'] - data['budget']

# AND ALSO CHECK THE NAMES OF TOP 10 PROFITABLE MOVIES

data[['Profit','movie_title']].sort_values(by = 'Profit', ascending = False).head(10)
```

```python
# WE ARE SEE THE RESULTS WE CAN EASILY ANALYZE THAT THERE ARE MANY SOME DUPLICATES

#LETS PRINT THE NO OF ROWS BEFORE REMOVING DUPLICATES
print("before removing duplicates: ",data.shape[0])

#REMOVE DUPLICATE VALUES
data.drop_duplicates(subset = None, keep = 'first', inplace = True)

#LETS PRINT AFTER THE REMOVING DUPLICATES
print("After Removing Duplicates: ",data.shape[0])
```

LETS CHECK TOP 10 PROFITABLE MOVIES AGAIN

```python
data[['Profit','movie_title']].sort_values(by = 'Profit', ascending = False).head(10)
```

```python
#LETS CHECK THE VALUE IN THE LANGUAGE COLUMNS
data['language'].value_counts()
```

```python
#WE ARE OBSERVE THAT OUT OF 3500 MOVIES ONLY 150 MOVIES ARE OF OTHER LANGUAGE
# SO IT IS EASIER TO KEEP ONLY 2 LANGUAGE THAT IS ENGLISH AND FOREIGN

def language(x):
    if x == 'English':
      return 'English'
    else:
      return 'Foreign'


# LETS APPLY THE FUNCTION ON THE LANGUAGE COLUMNS

data['language'] = data['language'].apply(language)

# LETS CHECK THE VALUE AGEIN
data['language'].value_counts()
```

```python
# LETS DEFINE A FUNCTION FOR CATEGORIZING DURATION OF LONG AND SHORT MOVIES


def duration(x):
  if x <= 120:
    return 'Short'
  else:
    return 'Long'


# LETS APPLY THIS FUNCTION ON THE DURATION COLUMNS

data['duration'] = data['duration'].apply(duration)

#LETS CHECK THE VALUES OF DURATION COLUMNS

data['duration'].value_counts()
```

```python
# LETS CHECK THE GENRES COLUMNS

data['genres'].value_counts()
```

```python
# THE ABOVE OUTPUT THAT MOST OF THE MOVIES ARE HAVING A LOT OF GENERES

# ALSO A MOVIE CAN HAVE SO MANY GENERES SO LETS KEEP FOUR GENERES


data['Moviegenres'] = data['genres'].str.split('|')
data['Genre1'] = data['Moviegenres'].apply(lambda x: x[0])

# SOME OF THE MOVIES HAVE ONLY ONE GENRE. IN CASE ASSIGN THE SAME GENRE TO "GENRE_2 AS WELL"

data['Genre2'] = data['Moviegenres'].apply(lambda x: x[1] if len(x) > 1 else x[0])
data['Genre3'] = data['Moviegenres'].apply(lambda x: x[2] if len(x) > 2 else x[0])
data['Genre4'] = data['Moviegenres'].apply(lambda x: x[3] if len(x) > 3 else x[0])



# Lets Check the Head of Data

data[['genres','Genre1','Genre2','Genre3','Genre4']].head(5)
```

```python
def query_actors(x):
  a= data[data['actor_1_name'] == x]
  b= data[data['actor_2_name'] == x]
  c= data[data['actor_3_name'] == x]
  df = pd.concat([a,b,c], ignore_index=True)

  y = df[['movie_title',
         'budget',
         'gross',
         'title_year',
         'genres',
         'language',
         'imdb_score',
         ]]
  return y
```

```python
def actors_report(x):
  a= data[data['actor_1_name'] == x]
  b= data[data['actor_2_name'] == x]
  c= data[data['actor_3_name'] == x]
  df = pd.concat([a,b,c], ignore_index=True)
  print("Time:",df['title_year'].min(), df['title_year'].max())
  print("Max Gross : {0:.2f} Millions".format(df['gross'].max()))
  print("Avg Gross : {0:.2f} Millions".format(df['gross'].mean()))
  print("Min Gross : {0:.2f} Millions".format(df['gross'].min()))
  print("No of  100 Millions Movies :", df[df['gross'] > 100].shape[0])
  print("AVG IMDB SCORE : {0:.2f}".format(df['imdb_score'].mean()))
  print("Most Common Genres:\n",df['Genre1'].value_counts().head())


actors_report('Tom Cruise')
```

```python
def critically_acclaimed_actors(m):
  a= data[data['actor_1_name'] == m]
  b= data[data['actor_2_name'] == m]
  c= data[data['actor_3_name'] == m]
  df = pd.concat([a,b,c], ignore_index=True)
  return df['num_critic_for_reviews'].sum().astype('int')


print("Number of Critics Reviews for Brad Pitt")
display(critically_acclaimed_actors('Brad Pitt'))


print("Number of Critics Reviews for Leonardo DiCaprio")
display(critically_acclaimed_actors('Leonardo DiCaprio'))


print("Number of Critics Reviews for Tom Cruise")
display(critically_acclaimed_actors('Tom Cruise'))
```

## **RECOMMENDING MOVIES BASED ON LANGUAGES**

## **RECOMMENDING MOVIES BASED ON ACTORS**

```python
def recommend_movies_on_actors(x):
  a = data[['movie_title', 'imdb_score']][data['actor_1_name'] == x]
  b = data[['movie_title', 'imdb_score']][data['actor_2_name'] == x]
  c = data[['movie_title', 'imdb_score']][data['actor_3_name'] == x]

  df = pd.concat([a,b,c], ignore_index= True)
  df= df.sort_values(by='imdb_score', ascending = False)
  return df.head(15)
```

# **Similar Genres Movies**

```python
from mlxtend.preprocessing import TransactionEncoder


x = data['genres'].str.split('|')
te = TransactionEncoder()
x = te.fit_transform(x)
x = pd.DataFrame(x, columns = te.columns_)

#WE are checking the head of X

x.head()
```

```python
# we are convert the data into boolean so that we can perform calculation
genres = x.astype('int')

genres.head()
```

```python
# now lets insert the movie titles in the first columns so that we can better understand data

genres.insert(0, 'movie_title', data['movie_title'])

genres.head()
```

```python
# we are set these movie titles as index of the data

genres =genres.set_index('movie_title')

genres.head()
```

```python
# making a recommendation engine for getting similar genres

def recommendation_genres(gen):
  gen= genres[gen]
  similar_genres = genres.corrwith(gen)
  similar_genres = similar_genres.sort_values(ascending=False)
  similar_genres = similar_genres.iloc[1:]
  return similar_genres.head(3)
```

```python
recommendation_genres('Action')
```

# **RECOMMENDING SIMILAR MOVIES**

```python
# we are making a sparse matrix to recommend the movies

x = genres.transpose()
x.head()
```

```python
# making a recoomendation engine for getting a similar movies


def recommendation_movie(movie):
  movie = x[movie+'\xa0']
  similar_movies = x.corrwith(movie)
  similar_movies = similar_movies.sort_values(ascending=False)
  similar_movies = similar_movies.iloc[1:]
  return similar_movies.head(20)
```

```python
recommendation_movie('The Expendables')
```
